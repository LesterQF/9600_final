## Data for this assessment

The data used for this assessment consist of records from Intensive Care Unit (ICU) hospital stays in the USA. All patients were adults who were admitted for a wide variety of reasons. ICU stays of less than 48 hours have been excluded.

The source data for the assessment are data made freely available for the 2012 MIT PhysioNet/Computing for Cardiology Challenge. Details are provided [here](https://physionet.org/challenge/2012/). Training Set A data have been used. The original data has been modified and assembled to suit the purpose of this assessment. Full details of the preparatory work is not required for the purposes of this assignment.

The dataframe consists of 120 variables, which are defined as follows:

```r
install.packages("glmnet")
library(glmnet)

install.packages("mice")
library(mice)

install.packages("dplyr")
library(dplyr)
```

```r
# See the percentage of missing variables
missing_percentage <- colMeans(is.na(icu_patients_df1)) * 100

# Output the result
print(missing_percentage)
```

```r
dataset<-icu_patients_df1
threshold <- 50  # Set a threshold, in percentage terms

# Calculate the proportion of missing values for each column
missing_percent <- colMeans(is.na(dataset)) * 100
print(missing_percent)
# Process according to threshold
selected_columns <- names(dataset)[missing_percent < threshold]
deleted_columns <- names(dataset)[missing_percent >= threshold]

# Delete columns that are larger than the threshold
cleaned_dataset <- dataset[, selected_columns]

# Median interpolation for columns that are less than the threshold
for (col in names(cleaned_dataset)) {
  if (any(is.na(cleaned_dataset[[col]]))) {
    median_value <- median(cleaned_dataset[[col]], na.rm = TRUE)
    cleaned_dataset[[col]][is.na(cleaned_dataset[[col]])] <- median_value
  }
}

# Output the processed data set
head(cleaned_dataset)
```

```r
# Columns that require one-hot coding
encoded_gender <- model.matrix(~ Gender - 1, data = cleaned_dataset)
encoded_icu_type <- model.matrix(~ ICUType - 1, data = cleaned_dataset)
encoded_status_type <- model.matrix(~ Status - 1, data = cleaned_dataset)

# Merge encoded data
encoded_dataset <- cbind(cleaned_dataset, encoded_gender,encoded_icu_type,encoded_status_type)

# Delete the original categorical column
encoded_dataset <- encoded_dataset[, !colnames(encoded_dataset) %in% c("Gender", "ICUType")]

# Output the processed data set
head(encoded_dataset)








```

```r
# Delete the "RecordID" variable
encoded_dataset <- encoded_dataset[, !(colnames(encoded_dataset) %in% c("RecordID"))]

# Separate target variables from predictor variables
target_variable <- encoded_dataset$in_hospital_death
predictor_variables <- encoded_dataset[, !(colnames(encoded_dataset) %in% c("in_hospital_death"))]

# Standardize predictor variables
scaled_predictor_variables <- scale(predictor_variables)

#LASSO regression
lasso_fit <- cv.glmnet(scaled_predictor_variables, target_variable, alpha = 1)

# Extract variables with non-zero coefficients
lasso_selected_vars <- as.vector(coef(lasso_fit, s = "lambda.min")[-1,])
selected_vars <- colnames(predictor_variables)[lasso_selected_vars != 0]

# View coefficient values for the LASSO regression model
lasso_coefficients <- coef(lasso_fit, s = "lambda.min")

# Calculate the absolute value of the coefficient value
lasso_coeff_abs <- abs(lasso_coefficients)

# Convert a sparse matrix to a normal matrix
lasso_coeff_matrix <- as.matrix(lasso_coeff_abs)

# Convert the absolute value part of the matrix to a vector
lasso_coeff_vector <- as.vector(lasso_coeff_matrix)

# Create a data box containing the variable name and the corresponding absolute value
coeff_abs_df <- data.frame(variable = rownames(lasso_coeff_matrix), absolute_value = lasso_coeff_vector)

# Sort from largest to smallest in absolute value
sorted_coeff_abs_df <- coeff_abs_df[order(coeff_abs_df$absolute_value, decreasing = TRUE), ]

# Outputs the sorted variable name and absolute value
print(sorted_coeff_abs_df)
```

```r
# Construct logistic regression model
model_1 <- glm(in_hospital_death ~ GCS_max + SOFA + BUN_min , data = encoded_dataset, family = "binomial")
summary(model_1)

model_2 <- glm(in_hospital_death ~ GCS_max + SOFA + BUN_min + DiasABP_max , data = encoded_dataset, family = "binomial")
summary(model_2)

model_3 <- glm(in_hospital_death ~ GCS_max + SOFA + BUN_min + DiasABP_max + Length_of_stay , data = encoded_dataset, family = "binomial")
summary(model_3)

model_4 <- glm(in_hospital_death ~ GCS_max + SOFA + BUN_min + DiasABP_max + Length_of_stay + Bilirubin_min , data = encoded_dataset, family = "binomial")
summary(model_4)
```
